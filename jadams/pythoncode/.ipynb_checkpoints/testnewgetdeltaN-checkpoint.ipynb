```python
import numpy as np
import matplotlib.pyplot as plt
import diffusionstuff as ds; reload(ds)
%matplotlib inline
```

```python
Nice = np.linspace(900,901)
Nbar = 1
Nstar = .111
Nqll = Nbar + Nstar*np.sin(2*np.pi*Nice)
plt.plot(Nice,Nqll)
plt.xlabel('$N_{ice}$')
plt.ylabel('$N_{QLL}$')
plt.grid('on')
```

```python
Ntot = Nice + Nqll
plt.plot(Ntot,Nqll)
plt.xlabel('$N_{tot}$')
plt.ylabel('$N_{QLL}$')
plt.grid('on')
```

```python
def fqll_next(fqll_last,Ntot,Nstar,Nbar):
    fstar = Nstar/Nbar
    return 1 + fstar*np.sin(2*np.pi*(Ntot-Nbar*fqll_last))
```

```python
fqll = Nqll/Nbar
fqll_0 = 1.0
fqll_1 = fqll_next(fqll_0,Ntot,Nstar,Nbar)
fqll_2 = fqll_next(fqll_1,Ntot,Nstar,Nbar)
plt.plot(Ntot,Nqll,Ntot,fqll_1,Ntot,fqll_2)
plt.xlabel('$N_{tot}$')
plt.ylabel('$N_{QLL}$')
plt.grid('on')
```

```python
def Nqll_iter(Ntot,Nstar,Nbar):
    fqll_last = 1.0
    n = 7
    for i in range(n):
        fqll_last = fqll_next(fqll_last,Ntot,Nstar,Nbar)
    return fqll_last*Nbar
```

```python
Nqll_iterative = Nqll_iter(Ntot,Nstar,Nbar)
Ntot_onepoint = 1.5
Nqll_onepoint = Nqll_iter(Ntot_onepoint,Nstar,Nbar)
plt.plot(Ntot,Nqll,Ntot,Nqll_iterative,Ntot_onepoint,Nqll_onepoint,'o')
plt.xlabel('$N_{tot}$')
plt.ylabel('$N_{QLL}$')
plt.grid('on')
```
